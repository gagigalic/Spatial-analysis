```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="app", timeout=None)
```

```python
lat = []
lon = []

for name in loc_rest["Name"]:
    location = geolocator.geocode(name)
    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
        
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
        
```

```python
print(lat)
```

```python
print(lon)
```

```python
loc_rest["lat"] = lat
loc_rest["lon"] = lon
```
